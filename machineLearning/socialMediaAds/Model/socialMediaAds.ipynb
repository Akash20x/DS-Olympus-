```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix, accuracy_score
```

```python
data = pd.read_csv("E:\github\DS-Olympus-\machineLearning\socialMediaAds\Dataset\Social_Network_Ads.csv")
data.head(10)
```

```python
data.info()
print("Shape: ",data.shape)
```

```python
data.drop("User ID", axis=1,inplace = True)
data
```

```python
# Are there any null values????
data.isnull().sum()
```

```python
# features and target
features = data.drop("Purchased", axis=1)
target = data["Purchased"]
print("------------------FEATURES--------------------")
print(features)
print(' ')
print("--------------TARGET-----------------")
print(target)

```

```python
# handling categorical values
new_features = pd.get_dummies(features, drop_first = True)
print(new_features)
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_features,target,random_state=133)
```

```python
# model and fit
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
# predictions
y_pred = model.predict(x_test)
print("---------------------------CLASSIFICATION REPORT----------------------------------")
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
train_score = model.score(x_train,y_train)
test_score = model.score(x_test,y_test)
print("Training score:", round(train_score,2))
print("Testing score: ", test_score)
```

```python
cm = confusion_matrix(y_test,y_pred)
plot_confusion_matrix(model,x_test,y_test)
plt.show()
```
